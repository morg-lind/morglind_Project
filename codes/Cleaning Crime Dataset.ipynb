```python
# import pandas library
import pandas as pd
```

```python
# Made a table in excel from original data and filtered for years 2024 and 2025. Turn those results into a csv
# file and read the 2024 and 2025 data 
```

```python
# read in the csv file as pandas dataframe
crime_data = pd.read_csv('crime_data_24_25.csv', encoding="utf-8", sep=",")

# display 
crime_data
```

```python
# display the data types
crime_data.dtypes
```

```python
# display the shape
crime_data.shape
```

```python
# check for duplicates in the dataframe
duplicates = crime_data[crime_data.duplicated(keep=False)]


print(f"total records: {len(crime_data)} \n")
print(f"duplicate records: {len(crime_data[crime_data.duplicated(keep='first')])} \n")
display(duplicates)
```

```python
# check for near duplicates using the DR_NO column
duplicate_numbers = crime_data[crime_data['DR_NO'].duplicated(keep=False)]

display(duplicate_numbers)
```

```python
# check total count of missing data
print(f"Total missing data: {crime_data.isna().sum().sum()}")
print("\n")

# check total count of missing data per column
print("Missing data per column: \n", "-"*50 )
print(crime_data.isna().sum())
```

```python
# replace null values in VICT_SEX and VICT_DESCENT with 'X' (unknown information)
crime_data['VICT_SEX'] = crime_data['VICT_SEX'].fillna('X')
crime_data['VICT_DESCENT'] = crime_data['VICT_DESCENT'].fillna('X')

# replace null values in MOCODES column with 'Unknown'
crime_data['MOCODES'] = crime_data['MOCODES'].fillna('Unknown')
```

```python
# replace null values in WEAPON_USED_CD and WEAPON_DESC with 'Unknown / None Used' (either unknown weapon used or no weapon used)
crime_data['WEAPON_USED_CD'] = crime_data['WEAPON_USED_CD'].fillna('Unknown / None Used')
crime_data['WEAPON_DESC'] = crime_data['WEAPON_DESC'].fillna('Unknown / None Used')
```

```python
# check total count of missing data per column
print("Missing data per column: \n", "-"*50 )
print(crime_data.isna().sum())
```

```python
# replace the age of 0 in VICT_AGE column with 'X' 
crime_data['VICT_AGE'] = crime_data['VICT_AGE'].replace(0, 'X')
```

```python
# display unique values in STATUS column
print(crime_data["STATUS"].unique())
```

```python
# display unique values in STATUS column
print(crime_data["STATUS_DESC"].unique())
```

```python
# find instances in STATUS column where values are null
crime_data_status = crime_data[crime_data['STATUS'].isna()]

# display those values
crime_data_status
```

The corresponding STATUS code is 'CC'.

```python
# replace the STATUS null value with 'CC'
crime_data['STATUS'] = crime_data['STATUS'].fillna('CC')
```

```python
# check total count of missing data per column to ensure the data cleaning is accurate
print("Missing data per column: \n", "-"*50 )
print(crime_data.isna().sum())
```
